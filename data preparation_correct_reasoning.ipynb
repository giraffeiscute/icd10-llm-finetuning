```python
import pandas as pd

df = pd.read_csv('mimiciv_note_4000_v2.csv', encoding='utf-8')

# 使用 extract_hpi_section 函數處理所有記錄
def process_records(df):
    # 建立一個新的 DataFrame 以保存處理後的結果
    processed_df = df.copy()
    # 儲存要刪除的索引
    rows_to_drop = []
    
    # 處理每一筆記錄
    for i in range(len(df)):
        text = df.iloc[i, 1]  # 使用 iloc 正確存取第二欄的內容
        processed_text = extract_hpi_section(text)  # 使用函數處理文字
        if processed_text:  # 如果有返回結果
            processed_df.iloc[i, 1] = processed_text  # 替換原本的文字
        else:  # 如果沒有找到 History of Present Illness
            rows_to_drop.append(i)
        
        # 每處理 100 筆記錄即顯示進度
        if (i + 1) % 100 == 0:
            print(f'已處理 {i + 1} 筆記錄')
    
    # 刪除沒有找到 History of Present Illness 的記錄
    processed_df = processed_df.drop(rows_to_drop)
    # 重置索引
    processed_df = processed_df.reset_index(drop=True)
    
    # 顯示刪除的記錄數
    if rows_to_drop:
        print(f'刪除了 {len(rows_to_drop)} 筆沒有 History of Present Illness 的記錄')
    
    return processed_df

# 執行處理
processed_df = process_records(df)

# 顯示結果
print('處理完成！')
print(f'處理後剩下 {len(processed_df)} 筆記錄')
```

# 製作成 LLM可讀之input json 檔案

```python
import pandas as pd
import json

# 讀取 CSV 檔案
df = pd.read_csv('mimiciv_note_4000_extracted.csv')

note_icd_data = []

# 遍歷每一筆資料
for idx, row in df.iterrows():
    hadm_id = row['hadm_id']
    note_text = row['note_text']
    icd_codes = row['icd_codes']
    
    # 如果病歷文字為空或是 NaN，跳過這筆資料
    if pd.isna(note_text) or not str(note_text).strip():
        continue
    
    # 處理 ICD 編碼
    if pd.isna(icd_codes) or not str(icd_codes).strip():
        icds = '無診斷紀錄'
    else:
        icds = str(icd_codes)
    
    # 確保 note_text 是字串
    note = str(note_text)
    
    question = f"There is a patient with the following ICD-10 codes:：{icds}. The medical record for the Patient is as follows:{note}。Please help me, using the medical record above, infer why each of this patient's ICD-10 codes was assigned. Please respond in the specified format."
    answer = f"{icds}。"
    
    note_icd_data.append({
        "question": question,
        "answer": answer
    })

# 儲存為 JSONL 格式
with open("note_icd_data_with_ans.jsonl", "w", encoding="utf-8") as f:
    for item in note_icd_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(f"\n成功處理 {len(note_icd_data)} 筆資料")
print(f"已儲存至 patient_icd_data.jsonl")
```

# 測試LLM input


```python
def extract_icd_answer(text: str) -> str:
    """
    從模型的完整回覆或標準答案中提取 ICD-10 編碼。
    優先從 <answer> 標籤中提取,若無則從整個文本中提取。
    """
    # 使用正則表達式搜尋 <answer> 標籤，re.DOTALL 讓 . 可匹配換行符，re.IGNORECASE 忽略大小寫
    answer_match = re.search(r'<answer>(.*?)</answer>', text, re.DOTALL | re.IGNORECASE)
    # 如果找到 <answer> 標籤，則只在標籤內容中搜尋 ICD 編碼；否則在整個文本中搜尋
    search_text = answer_match.group(1) if answer_match else text
    
    # 匹配格式：字母 + 1-3位數字 + 可選的小數點和1-2位數字
    codes_list = re.findall(r'\b([A-Z]\d{1,6}(?:\.\d{1,2})?)\b', search_text, re.IGNORECASE)
    
    # 移除小數點、去重、排序、轉大寫，並過濾掉長度<=2的編碼
    cleaned_codes = sorted(set(
        code.replace('.', '').upper() 
        for code in codes_list 
        if len(code.replace('.', '')) > 2
    ))
    return " ".join(cleaned_codes)
```

```python
import json
import re

chat_style_data = []

with open("note_icd_data.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)  
        
        chat_style_data.append({
            "prompt": [
                {"role": "user", "content": item["question"]}
            ],
            "answer": extract_icd_answer(item["answer"])  # ground truth ICD-10 codes
        })

print(f"總共處理了 {len(chat_style_data)} 筆資料")

```
